```python
import re;
#myfile = open('stock.txt')
while(True):
    try:
        fname = input('Give the name of the movie reviews file: ')
        myfile = open(fname)
        break;
    except:
        print('File cannot be opened:', fname)

print('Apple inc. stock prices:')

for line in myfile:
    stockPrice = re.findall('[0-9]+\.[0-9]+',line)
    stockDate = re.findall('[A-Z][a-z][a-z]\s[0-9]+',line)
    stockDateSplit = stockDate[0].split()
    if(len(stockDateSplit[1]) == 1):
        print(str(stockDateSplit[0])+" 0"+str(stockDateSplit[1])+", $"+str(stockPrice[0]))
    else:
        print(str(stockDateSplit[0])+" "+str(stockDateSplit[1])+", $"+str(stockPrice[0]))
```

## Question 2
Logic:
1. Added infinite loop to open the file until the file name is correct
2. Traversing each line in the file and replacing \n to empty. Spliting the line with ,.
3. Based on the list traversing from 1 to len, cause in index 0 we have reviewer name. So we are taking even index to rating list and odd index to name list. Then we are adding respecting movie and rating into dictionary. Finaly we are adding 0 index and movie list dict to movie dict.
4. inpCheck() - we are getting input from user. It should be numbers and the numbers should be either 1 or 2 or 3. If user enter anything other 1 or 2 or 3, it will run a infinite loop. If user enter correct input either 1 or 2 it will review() function
5. review() - we are getting reviewer name. If the choice is 1 we are going to get 2 reviewers, if the choice is 2, we are going to get 1 reviewer name. If the reviewers name is not in the dictionary, it will return error and run an infinite loop. If the user enter correct reviewer name it will call cosine method. If the user enters 2 reviewers name the it will call cosineSocre() function. If the user enter 1 reviewer name, then it will call cosineSocreForAll() function.
6. cosineSocre() - we are passing our main dictionary and two reviers name. First we need to get the sub dict for each reviewer. Iterating nested for loop for each reviewer sub dict. if the keys of both sub dict is same then will square the values of respective keys and will make sum of all squares. Finally will return the square root of the sum of all squares.
7. cosineSocreForAll() - it is similar to cosineSocre() function but here we are comparing with all other reviewers. We are passing our main dict and one reviewer name. First we are getting values for the given reviewer. Then we are iterating each key and values in the main dict and comparing the reviewer with user input reviewer name. If it not equal, we are initiating nested for loop with each values of sub dict and user reviewer values.Finally we adding cosine score, reviewer name and comparing reviewer name into the list and returning it.

```python
import math

while(True):
    try:
        fname = input('Give the name of the movie reviews file: ')
        myfile = open(fname)
        break;
    except:
        print('File cannot be opened:', fname)
#myfile = open('movie_reviews.txt')

movieDict = {}
for line in myfile:
    line = line.replace('\n', '')
    movieList=line.split(',')
    movieListDict = {}
    movieNameList = []
    movieRatingList = []
    for i in range(1, len(movieList)):
        if(i % 2 == 0):
            movieRatingList.append(float(movieList[i]))
        else:
            movieNameList.append(str(movieList[i]))
    for j in range(0,len(movieNameList)):
        if(len(movieNameList[j]) > 0 and len(movieRatingList) > 0):
            movieListDict[movieNameList[j]] = movieRatingList[j]
    if(len(movieList[0]) > 0):
        movieDict[movieList[0]] = movieListDict
print(movieDict)

def cosineSocre(movieDictionary, rev1, rev2):
    cosSocre = 0
    rev1Dic = movieDictionary[rev1]
    rev2Dic = movieDictionary[rev2]
    for r1key,r1val in rev1Dic.items():
        for r2key,r2val in rev2Dic.items():
            if(r1key == r2key):
                cosSocre = cosSocre+((r1val-r2val)**2)
    return math.sqrt(cosSocre)

def cosineSocreForAll(movieDictionary, rev1):
    rev1Dic = movieDictionary[rev1]
    cosMainList = []
    for r1key,r1val in movieDictionary.items():
        if(r1key != rev1):
            r1subDict = movieDictionary[r1key]
            cosSocre = 0
            cossqrt = 0;
            cosList = []
            for r1subKey, r1subVal in r1subDict.items():
                for r2key,r2val in rev1Dic.items():
                    if(r1subKey == r2key):
                        cosSocre = cosSocre+((r1subVal-r2val)**2)
            cossqrt = math.sqrt(cosSocre)
            cosList.append(rev1)
            cosList.append(r1key)
            cosList.append(cossqrt)
            cosMainList.append(cosList)    
    return cosMainList


def review(inpChoice):
    if(inpChoice == 1):
        while(True):
            inputReviewer1 = input("Provide Reviewer 1 name: ")
            inputReviewer2 = input("Provide Reviewer 2 name: ")
            if (inputReviewer1 in movieDict):
                if(inputReviewer2 in movieDict):
                    cosScore = cosineSocre(movieDict, inputReviewer1, inputReviewer2)
                    print("The similarity score between " +inputReviewer1+ " and " +inputReviewer2+ " is: "+str(round(cosScore,4)))
                    break
                else:
                    print("Reviewer 2 name is not present in the file")
            else:
                print("Reviewer 1 name is not present in the file")
    elif(inpChoice == 2):
        while(True):
            inputReviewer = input("Provide Reviewer name: ")
            if(inputReviewer in movieDict):
                codScoreAll = cosineSocreForAll(movieDict, inputReviewer)
                print("The Similarity Scores are: ")
                for i in codScoreAll:
                    print(str(i[0])+ "\t"+str(i[1])+"\t"+str(round(i[2],2)))
                break
            else:
                print("Reviewer name is not present in the file")
        
        
def inp_check(inputStr):
    booleanCheck = True;
    while(booleanCheck):
        try:
            inp = int(input(inputStr))
            if(inp == 1 or inp == 2):
                review(inp)
            elif(inp ==3):
                print("GoodBye!")
                break;
            else:
                print('Error: Enter choice 1 or 2 or 3')
        except:
            print('Error: You didn’t enter number')

inputChoice = inp_check("What do you want to do? Input 1 for similarity between two reviewers, or Input 2 for similarity between one reviewer and all others in the database or 3 to quit: ")

```
